```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Data Gaikindo Update.csv')
data.head()
```

```python
data.replace('-', 0, inplace=True)
data.head()
```

```python
data = data.fillna(0)
data.isnull().sum()
```

```python
numeric_cols = ['Total_CKD_Set_Export', 'Total_Component_Part_Export', 'Total_CBU_Import', 'Total_CKD_Set_Import',
                'Total_Component_Part_Import', 'Total_Production']
data[numeric_cols] = data[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
data.describe()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='Year', y='Total_Production', data=data, marker='o')
plt.title('Total Production Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Year', y='Total_Production', data=data)
plt.title('Total Production Each Year')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Year', y='Total_Production', data=data)
plt.title('Box Plot of Total Production Each Year')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.show()
```
